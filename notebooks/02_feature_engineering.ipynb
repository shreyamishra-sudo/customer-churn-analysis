```python
import pandas as pd

df = pd.read_csv(
    "../data/cleaned/cleaned_customer_churn.csv"
)

df.head()
```

```python
df["tenure_group"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 24, 48, 60, 72],
    labels=[
        "0-1 Year",
        "1-2 Years",
        "2-4 Years",
        "4-5 Years",
        "5-6 Years"
    ]
)

df["tenure_group"].value_counts()
```

```python
services = [
    "PhoneService",
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies"
]

df["total_services"] = df[services].apply(
    lambda x: (x == "Yes").sum(),
    axis=1
)

df["total_services"].head()
```

```python
df["monthly_charge_group"] = pd.cut(
    df["MonthlyCharges"],
    bins=[0, 35, 70, 100, 120],
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
)

df["monthly_charge_group"].value_counts()
```

```python
df["avg_monthly_spend"] = (
    df["TotalCharges"] /
    (df["tenure"] + 1)
)

df["avg_monthly_spend"].head()
```

```python
df.to_csv(
    "../data/cleaned/cleaned_customer_churn.csv",
    index=False
)
```
