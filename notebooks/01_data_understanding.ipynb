```python
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)
```

```python
df = df.dropna()
```

```python
df = df.drop("customerID", axis=1)
```

```python
df["Churn"] = df["Churn"].map({
    "Yes": 1,
    "No": 0
})
```

```python
df.to_csv(
    "../data/cleaned/cleaned_customer_churn.csv",
    index=False
)
```
